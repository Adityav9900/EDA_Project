Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

Loading the dataset

```python
df = pd.read_csv('Diwali_Sales_Data.csv',encoding = 'unicode_escape')
```

Knowing our Dataset

```python
df.shape
df.head()
```

```python
df.info()
```

Data Cleaning

```python
df.drop(['Status','unnamed1'],axis = 1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#Removing NULL values
df.dropna(inplace=True)
df.shape
```

```python
df.info()
```

```python
#Changing the data type
df['Amount'] = df['Amount'].astype(int)
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

Exploratory Data Analysis(EDA)

```python
df.columns
```

**Gender**

```python
ax = sb.countplot(x = 'Gender',data = df)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x='Gender', y = 'Amount',data = sales_gen)
```

**Age**

```python
ax = sb.countplot(data = df, x = 'Age Group',hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Total amount vs Age Group
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x = 'Age Group',y = 'Amount',data = sales_age)
```

```python
#Statewise orders
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sb.set(rc = {'figure.figsize':(20,5)})
sb.barplot(x = 'State', y = 'Orders',data = sales_state)
```

```python
#Statewise amount of purchase
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sb.set(rc = {'figure.figsize':(20,5)})
sb.barplot(x = 'State', y = 'Amount',data = sales_state)
```

Marital **Status**

```python
ax = sb.countplot(x = 'Marital_Status',data = df)
sb.set(rc = {'figure.figsize':(3,1)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_mar = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sb.set(rc = {'figure.figsize':(10,5)})
sb.barplot(x = 'Marital_Status', y = 'Amount',data = sales_mar,hue='Gender')
```

**Occupation**

```python
ax = sb.countplot(x = 'Occupation',data = df)
sb.set(rc = {'figure.figsize':(25,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sb.barplot(x = 'Occupation', y = 'Amount',data = sales_occ)
```

Product **Category**

```python
ax = sb.countplot(x = 'Product_Category',data = df)
sb.set(rc = {'figure.figsize':(25,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_cat = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(12)
sb.barplot(x = 'Product_Category', y = 'Amount',data = sales_cat)
```

```python
sales_cat = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
sb.barplot(x = 'Product_ID', y = 'Orders',data = sales_cat)
```
